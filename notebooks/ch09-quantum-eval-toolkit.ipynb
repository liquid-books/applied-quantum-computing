## Module 1: TCO Analyzer (Chapter 3 Framework)

Estimates the total cost of ownership for a quantum pilot versus its classical alternative.
Based on the framework from Chapter 3: *The New Economics of Computation.*


```python
# ============================================================
# MODULE 1: TCO ANALYZER
# ============================================================

## INPUTS — Edit these for your use case

# Quantum cloud costs
quantum_shots_per_run = 8192          # shots per circuit execution
quantum_runs_per_day = 10             # how many runs per working day
quantum_cost_per_shot = 0.00035       # USD per shot (IBM Quantum approximate)
quantum_team_fte = 1.5                # FTE equivalent working on quantum
quantum_fte_annual_cost = 220000      # USD per FTE per year (fully loaded)
quantum_pilot_duration_months = 12   # pilot duration in months
quantum_overhead_percent = 0.20       # overhead: tooling, training, admin

# Classical alternative costs
classical_cloud_monthly = 8000        # USD/month for equivalent classical HPC
classical_team_fte = 1.0              # FTE on classical solution
classical_fte_annual_cost = 160000    # USD per FTE per year

# Performance estimates
classical_baseline_quality = 100      # index: classical performance = 100
quantum_expected_quality = 105        # expected quantum performance index
business_value_per_quality_point = 50000  # USD per quality index point per year

## CALCULATIONS

working_days = quantum_pilot_duration_months * 21  # ~21 working days/month

# Quantum costs
q_compute = quantum_shots_per_run * quantum_runs_per_day * working_days * quantum_cost_per_shot
q_labor = quantum_team_fte * quantum_fte_annual_cost * (quantum_pilot_duration_months / 12)
q_subtotal = q_compute + q_labor
q_overhead = q_subtotal * quantum_overhead_percent
q_total = q_subtotal + q_overhead

# Classical costs
c_compute = classical_cloud_monthly * quantum_pilot_duration_months
c_labor = classical_team_fte * classical_fte_annual_cost * (quantum_pilot_duration_months / 12)
c_total = c_compute + c_labor

# Value analysis
quality_improvement = quantum_expected_quality - classical_baseline_quality
annual_value = quality_improvement * business_value_per_quality_point
incremental_cost = q_total - c_total
payback_years = incremental_cost / annual_value if annual_value > 0 else float('inf')

## OUTPUT

print("\n" + "="*60)
print("MODULE 1: TCO ANALYSIS")
print("="*60)

tco_table = [
    ["Quantum compute cost", f"${q_compute:,.0f}"],
    ["Quantum labor cost", f"${q_labor:,.0f}"],
    ["Quantum overhead", f"${q_overhead:,.0f}"],
    ["QUANTUM TOTAL", f"${q_total:,.0f}"],
    ["", ""],
    ["Classical compute cost", f"${c_compute:,.0f}"],
    ["Classical labor cost", f"${c_labor:,.0f}"],
    ["CLASSICAL TOTAL", f"${c_total:,.0f}"],
    ["", ""],
    ["Incremental quantum cost", f"${incremental_cost:,.0f}"],
    ["Expected quality improvement", f"{quality_improvement} index points"],
    ["Annual business value", f"${annual_value:,.0f}"],
    ["Payback period", f"{payback_years:.1f} years" if payback_years != float('inf') else "N/A"]
]

print(tabulate(tco_table, headers=["Item", "Value"], tablefmt="rounded_outline"))

if payback_years < 3:
    verdict = "✅ FAVORABLE — payback within 3 years"
elif payback_years < 7:
    verdict = "⚠️ MARGINAL — long payback; option value may justify"
else:
    verdict = "❌ UNFAVORABLE — payback too long; reconsider or defer"

print(f"\nTCO Verdict: {verdict}")

# Store for summary report
tco_result = {
    'quantum_total': q_total,
    'classical_total': c_total,
    'incremental_cost': incremental_cost,
    'annual_value': annual_value,
    'payback_years': payback_years,
    'verdict': verdict
}
```

## Module 2: Hardware Rubric Scorer (Chapter 4 Framework)

Scores multiple quantum hardware platforms against your use case requirements.
Based on the hardware evaluation rubric from Chapter 4: *The Quantum Hardware Race.*


```python
# ============================================================
# MODULE 2: HARDWARE RUBRIC SCORER
# ============================================================

## INPUTS — Edit these for your use case

# Your requirements (0-10 scale: 10 = absolutely critical, 0 = irrelevant)
requirements = {
    'qubit_count':          9,   # How many physical qubits do you need?
    'gate_fidelity':        8,   # How error-sensitive is your algorithm?
    'connectivity':         7,   # Do you need all-to-all qubit connectivity?
    'coherence_time':       7,   # How deep are your circuits (long coherence needed)?
    'cloud_accessibility':  9,   # Is cloud access required (vs. on-premises)?
    'sdk_ecosystem':        8,   # How important is a mature software ecosystem?
    'cost_per_shot':        6,   # How cost-sensitive is your budget?
    'error_correction':     5,   # Do you need quantum error correction support?
}

# Platform specifications — 2025 approximate values
# Format: {requirement: score_0_to_10}
platforms = {
    'IBM Quantum (Eagle/Heron)': {
        'qubit_count': 9,       # 127-433 qubits available
        'gate_fidelity': 8,     # 99.5%+ 2Q gate fidelity on best devices
        'connectivity': 6,      # Heavy-hex connectivity (not all-to-all)
        'coherence_time': 7,    # ~300-500 microseconds T2
        'cloud_accessibility': 10,  # Best-in-class cloud access
        'sdk_ecosystem': 10,    # Qiskit is industry standard
        'cost_per_shot': 6,     # Competitive but not cheapest
        'error_correction': 7,  # Active QEC research; not production yet
    },
    'IonQ (Forte)': {
        'qubit_count': 7,       # 35 algorithmic qubits (higher quality)
        'gate_fidelity': 10,    # 99.9%+ 2Q gate fidelity
        'connectivity': 10,     # All-to-all connectivity
        'coherence_time': 9,    # Trapped ions: seconds of coherence
        'cloud_accessibility': 8,   # Available via AWS Braket, Azure Quantum
        'sdk_ecosystem': 7,     # Good but smaller than Qiskit
        'cost_per_shot': 3,     # More expensive per shot
        'error_correction': 5,  # Less focus on QEC currently
    },
    'Google Quantum AI': {
        'qubit_count': 9,       # 70+ qubits on Sycamore
        'gate_fidelity': 9,     # State-of-the-art superconducting
        'connectivity': 7,      # Grid connectivity
        'coherence_time': 7,    # Similar to IBM superconducting
        'cloud_accessibility': 5,   # Limited cloud access (research partnerships)
        'sdk_ecosystem': 7,     # Cirq ecosystem, growing
        'cost_per_shot': 4,     # Not widely available commercially
        'error_correction': 10, # Leading QEC research (below-threshold demonstrated)
    },
    'AWS Braket (Multi-vendor)': {
        'qubit_count': 8,       # Access to multiple hardware backends
        'gate_fidelity': 7,     # Varies by backend
        'connectivity': 7,      # Varies by backend
        'coherence_time': 6,    # Varies by backend
        'cloud_accessibility': 10,  # Native AWS integration
        'sdk_ecosystem': 8,     # Good AWS integration, multi-backend support
        'cost_per_shot': 7,     # Transparent pricing
        'error_correction': 5,  # Dependent on backend
    }
}

## CALCULATIONS

total_weight = sum(requirements.values())

scores = {}
for platform, specs in platforms.items():
    weighted_score = sum(
        requirements[k] * specs.get(k, 5) for k in requirements
    )
    normalized_score = (weighted_score / (total_weight * 10)) * 100
    scores[platform] = round(normalized_score, 1)

ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)

## OUTPUT

print("\n" + "="*60)
print("MODULE 2: HARDWARE RUBRIC SCORES")
print("="*60)

rubric_table = [[rank+1, platform, f"{score}/100",
                 "✅ Recommended" if score >= 70 else
                 "⚠️ Consider" if score >= 50 else
                 "❌ Not suitable"]
                for rank, (platform, score) in enumerate(ranked)]

print(tabulate(rubric_table,
               headers=["Rank", "Platform", "Score", "Assessment"],
               tablefmt="rounded_outline"))

top_platform = ranked[0][0]
top_score = ranked[0][1]
print(f"\n🏆 Recommended platform: {top_platform} ({top_score}/100)")

# Store for summary
rubric_result = {
    'top_platform': top_platform,
    'top_score': top_score,
    'ranked': ranked
}
```

## Module 3: Mosca Inequality Classifier (Chapter 5 Framework)

Classifies PQC migration urgency using the Mosca Inequality: **(X + Y) vs Z**

- **X** = how many years your sensitive data must remain confidential
- **Y** = how many years your cryptographic migration will take
- **Z** = estimated years until a cryptographically relevant quantum computer exists

If **(X + Y) > Z** → you are already in the danger zone. Act now.


```python
# ============================================================
# MODULE 3: MOSCA INEQUALITY CLASSIFIER
# ============================================================

## INPUTS — Edit these for your organization

# X: How long does your most sensitive data need to remain confidential?
# (Consider: trade secrets, financial records, health records, national security)
X_data_sensitivity_years = 10  # years

# Y: How long will your cryptographic migration take?
# (Typical range: 3-7 years for large enterprises, 2-4 for smaller organizations)
Y_migration_years = 5  # years

# Z: When will a cryptographically relevant quantum computer exist?
# (Consensus range: 7-15 years; use a range for sensitivity analysis)
Z_threat_optimistic = 8   # years (optimistic: quantum arrives sooner)
Z_threat_consensus = 12   # years (consensus estimate)
Z_threat_pessimistic = 18 # years (pessimistic: quantum arrives later)

## CALCULATIONS

XY = X_data_sensitivity_years + Y_migration_years

def classify_mosca(xy, z, scenario):
    gap = xy - z
    if xy > z:
        status = "🔴 CRITICAL — Already in danger zone"
        action = "Begin PQC migration immediately. You are behind schedule."
        urgency = min(10, int(gap * 1.5) + 7)
    elif xy == z:
        status = "🟡 DANGER ZONE EDGE — At the threshold"
        action = "Begin PQC migration now. No margin for delay."
        urgency = 7
    elif (z - xy) <= 2:
        status = "🟡 WARNING — Margin is very thin"
        action = "Begin PQC migration within 6 months."
        urgency = 6
    elif (z - xy) <= 5:
        status = "🟠 MONITOR — Limited time remaining"
        action = "Begin PQC migration planning this year."
        urgency = 4
    else:
        status = "🟢 SAFE — Sufficient margin"
        action = "Monitor standards; begin planning in 2-3 years."
        urgency = 2
    return {'scenario': scenario, 'Z': z, 'gap': gap, 'status': status, 'action': action, 'urgency': urgency}

scenarios = [
    classify_mosca(XY, Z_threat_optimistic, "Optimistic (Q arrives early)"),
    classify_mosca(XY, Z_threat_consensus, "Consensus estimate"),
    classify_mosca(XY, Z_threat_pessimistic, "Pessimistic (Q arrives late)"),
]

## OUTPUT

print("\n" + "="*60)
print("MODULE 3: MOSCA INEQUALITY ANALYSIS")
print("="*60)
print(f"\n  Data sensitivity lifetime (X): {X_data_sensitivity_years} years")
print(f"  Migration timeline (Y):         {Y_migration_years} years")
print(f"  X + Y combined exposure:         {XY} years")
print()

mosca_table = [[s['scenario'], s['Z'], f"{XY} vs {s['Z']}",
                '>' if XY > s['Z'] else '=' if XY == s['Z'] else '<',
                s['status']]
               for s in scenarios]

print(tabulate(mosca_table,
               headers=["Scenario", "Z (years)", "X+Y vs Z", "Sign", "Classification"],
               tablefmt="rounded_outline"))

print("\n📋 Recommended Actions by Scenario:")
for s in scenarios:
    print(f"  [{s['scenario']}]: {s['action']}")

# Overall recommendation based on consensus
consensus = scenarios[1]
print(f"\n⚠️  Overall urgency score (consensus): {consensus['urgency']}/10")

# Store for summary
mosca_result = {
    'XY': XY,
    'scenarios': scenarios,
    'consensus': consensus
}
```
